### Line chart of DOAS share prices between 01.01.2021-31.12.2023

```python
# Load the data from the specified Excel file
file_path_doas = "Tez_Hisse_Files/Tez_hisse_doas.xlsx"
df = pd.read_excel(file_path_doas)

# Convert the 'Tarih' column to datetime format
df['Tarih'] = pd.to_datetime(df['Tarih'])

# Create a line chart for the share prices
plt.figure(figsize=(12, 6))
plt.plot(df['Tarih'], df['Şimdi'], label='DOAS Share Price', color='green',linestyle='--', linewidth=3)
#plt.title('DOAS Share Prices (01.01.2021 - 31.12.2023)')
plt.xlabel('Date')
plt.ylabel('Share Price (TRY)')
plt.grid(True)
plt.legend()
plt.show()  # Display the line chart
```

## Scatter Plot

```python
# Scatter plot to visualize the relationship between opening and closing prices
plt.figure(figsize=(10, 6))
plt.scatter(df['Açılış'], df['Şimdi'], color='b', alpha=0.4,marker='x')
plt.title('Scatter Plot of Opening vs. Closing Prices')
plt.xlabel('Opening Price (TRY)')
plt.ylabel('Closing Price (TRY)')
plt.grid(True)
plt.show()  # Display the scatter plot
```

## Box Plot

```python
# Box plot of the share prices to understand the spread and identify outliers
plt.figure(figsize=(10, 6))
plt.boxplot(df['Şimdi'], vert=True, patch_artist=True, boxprops=dict(facecolor='skyblue'))
#plt.title('Box Plot of DOAS Share Prices')
plt.ylabel('Share Price (TRY)')
plt.grid(True)
plt.show()  # Display the box plot
```

## Histogram for DOAS

```python
# Histogram of the share prices to examine the distribution
plt.figure(figsize=(10, 6))
plt.hist(df['Şimdi'], bins=20, color='lightgreen', edgecolor='black')
plt.title('Distribution of DOAS Share Prices')
plt.xlabel('Share Price (TRY)')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()  # Display the histogram
```

### Simulation a few Times for DOAS

```python
# Set random seed for reproducibility
np.random.seed(0)

# Define x-axis: days from 1 to 244
days = np.arange(1, 245)

# Generate random initial values between 200 and 300
initial_values = np.random.uniform(200, 300, 10)

# Create 10 simulations with random fluctuations
simulations = [
    initial_values[i] + np.cumsum(np.random.normal(0, 5, len(days))) for i in range(10)
]

# Calculate the average across all simulations
average = np.mean(simulations, axis=0)

# Plot all simulations with unique colors
plt.figure(figsize=(8, 6))
for i, sim in enumerate(simulations):
    plt.plot(days, sim, label=f"{i + 1}. Simulations")

# Plot the average line with a distinct color and thicker linewidth
plt.plot(days, average, label="Mean", linewidth=3, color="navy")

# Add axis labels and title
plt.xlabel("Days")
plt.ylabel("Share Price (TRY)")
plt.title("Share Price Simulations")

# Add legend with best placement
plt.legend(loc="best")

# Enable grid for better readability
plt.grid(True)

# Display the plot
plt.show()
```

### Boxplot

```python
# Create a box plot with all simulations
plt.figure(figsize=(10, 6))  # Adjust figure size
plt.boxplot(simulations, patch_artist=True, boxprops=dict(facecolor='blue'))  # Plot the box plot with a patch for each box

# Customize the box plot
plt.xlabel("Simulations", fontsize=12)
plt.ylabel("Share Price (TRY)", fontsize=12)
plt.title("Box Plot of Share Price Simulations", fontsize=14, fontweight='bold')

# Display the box plot
plt.show()
```

### Simulation a few Times for FROTO

```python
# Load the data from the Excel file
file_path_froto = "Tez_Hisse_Files/Tez_hisse_froto.xlsx"
df = pd.read_excel(file_path_froto)

# Define initial values for 10 simulations from a specific column
# This assumes there's a column with initial share prices or similar data
initial_values = df.iloc[0:10, 1].values  # Extract the first 10 values from the second column

# Create 10 simulations with cumulative sums and random fluctuations
days = np.arange(1, 245)  # Define 244 days for x-axis
simulations = [
    initial_values[i] + np.cumsum(np.random.normal(0, 5, len(days))) for i in range(10)
]

# Calculate the average across all simulations
average = np.mean(simulations, axis=0)

# Create a new figure for plotting
fig, ax = plt.subplots(figsize=(8, 6))

# Plot the box plots for simulations with default box properties
plt.boxplot(simulations, labels=[f"{i + 1}. Simulation" for i in range(10)], patch_artist=True)

# Add a horizontal line for the mean
ax.axhline(y=np.mean(average), color='r', linestyle='--', label='Mean')

# Rotate x-axis labels for better readability
plt.xticks(rotation=65)

# Customize the box plot with axis labels and title
plt.ylabel("Share Price (TRY)", fontsize=12)

# Add a legend to identify the mean line
ax.legend(loc="best")

# Display the box plot
plt.show()  # Display the plot with the corrected horizontal line for the mean
```
