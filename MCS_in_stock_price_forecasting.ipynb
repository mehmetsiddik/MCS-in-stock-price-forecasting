```python
# Load the Excel file
file_path = "Tez_Hisse_Files/Tez_hisse_doas.xlsx"
excel_data = pd.ExcelFile(file_path)

# Load the first sheet
df = excel_data.parse('Sayfa1')

# Extract observed prices
observed_prices = pd.to_numeric(df['Şimdi'], errors='coerce').dropna().values

# Simulate 10 random price paths for demonstration
np.random.seed(42)
days = len(observed_prices)
simulated_prices = np.array([observed_prices + np.random.normal(0, 10, days) for _ in range(10)])

# Calculate Mean Simulated Prices for each day
mean_simulated_prices = np.mean(simulated_prices, axis=0)

# Calculate MAPE
mape = np.mean(np.abs((observed_prices - mean_simulated_prices) / observed_prices)) * 100
mape_text = f"Mean Absolute Percentage Error (MAPE): {mape:.2f}%"

# Plot Observed vs Simulated Prices
plt.figure(figsize=(12, 6))
plt.plot(range(1, days + 1), observed_prices, label="Observed Prices", color="blue", linewidth=2)
plt.plot(range(1, days + 1), mean_simulated_prices, label="Mean Simulated Prices", color="orange", linestyle="--", linewidth=2)
plt.fill_between(range(1, days + 1), 
                 np.min(simulated_prices, axis=0), 
                 np.max(simulated_prices, axis=0), 
                 color="gray", alpha=0.3, label="Simulation Range")
plt.xlabel("Days")
plt.ylabel("Stock Prices")
#plt.title("Observed vs Simulated Stock Prices")
plt.legend()
plt.grid(True)
plt.text(5, max(observed_prices) - 18, mape_text, fontsize=12, color='red')
plt.show()
```

```python
# Load Historical Data (e.g., Actual Stock Prices)
historical_prices = df2['Şimdi'].values[:244]  # Assuming 'Şimdi' column has historical stock prices

# Define Simulation Parameters
np.random.seed(42)
days = np.arange(1, 245)  # 244 days
initial_price = np.mean(historical_prices)  # Start simulation at historical mean price
num_simulations = 10000

# Run 10,000 Simulations
simulations = [
    initial_price + np.cumsum(np.random.normal(0, 5, len(days))) for _ in range(num_simulations)
]

# Calculate Mean Simulated Prices Across All Simulations
mean_simulated_prices = np.mean(simulations, axis=0)

# --- Validation ---
# 1. Compare Summary Statistics
historical_mean = np.mean(historical_prices)
simulated_mean = np.mean(mean_simulated_prices)
historical_std = np.std(historical_prices)
simulated_std = np.std(mean_simulated_prices)

print(f"Historical Mean: {historical_mean:.2f}, Simulated Mean: {simulated_mean:.2f}")
print(f"Historical Std Dev: {historical_std:.2f}, Simulated Std Dev: {simulated_std:.2f}")

# 2. Kolmogorov-Smirnov Test (Distribution Comparison)
ks_stat, p_value = ks_2samp(historical_prices, mean_simulated_prices)
print(f"KS Statistic: {ks_stat:.4f}, p-value: {p_value:.4f}")

# Visualize Historical vs Simulated Distributions
plt.figure(figsize=(10, 6))
plt.hist(historical_prices, bins=20, alpha=0.7, label="Historical Prices", density=True, color='blue')
plt.hist(mean_simulated_prices, bins=20, alpha=0.7, label="Simulated Prices", density=True, color='orange')
plt.xlabel("Stock Price (TRY)")
plt.ylabel("Density")
plt.title("Comparison of Historical and Simulated Price Distributions")
plt.legend()
plt.grid(True)
plt.show()

# --- Approximate Bayesian Computation (ABC) ---
# Example: Calibrating the standard deviation parameter of the normal distribution
observed_summary = np.array([historical_mean, historical_std])  # Observed statistics

# Define the distance function
def distance(simulated_summary, observed_summary):
    return np.sum((simulated_summary - observed_summary) ** 2)

# ABC Procedure
param_values = np.linspace(1, 10, 100)  # Candidate standard deviations
distances = []

for param in param_values:
    # Simulate data using candidate parameters
    simulations = [
        initial_price + np.cumsum(np.random.normal(0, param, len(days))) for _ in range(100)
    ]
    simulated_mean = np.mean([np.mean(sim) for sim in simulations])
    simulated_std = np.mean([np.std(sim) for sim in simulations])
    simulated_summary = np.array([simulated_mean, simulated_std])
    
    # Compute distance
    dist = distance(simulated_summary, observed_summary)
    distances.append(dist)

# Select the parameter with the smallest distance
best_param = param_values[np.argmin(distances)]
print(f"Optimal Parameter (Standard Deviation): {best_param:.2f}")

# Visualize the ABC Results
plt.figure(figsize=(10, 6))
plt.plot(param_values, distances, marker='o')
plt.xlabel("Candidate Parameter (Std Dev)")
plt.ylabel("Distance")
plt.title("ABC Calibration Results")
plt.grid(True)
plt.show()
```

```python
# Plot for ABC Calibration
plt.figure(figsize=(10, 6))
plt.plot(param_values, distances, marker='o', color='blue', label='Distance')
plt.axvline(best_param, color='red', linestyle='--', label=f'Optimal Std Dev = {best_param:.2f}')

# Adding bold text
plt.xlabel("Candidate Parameter (Standard Deviation)", fontsize=12, fontweight='bold')
plt.ylabel("Distance", fontsize=12, fontweight='bold')
plt.title("Approximate Bayesian Computation (ABC) Calibration", fontsize=14, fontweight='bold')
plt.legend(fontsize=10)
plt.grid(True)
plt.tight_layout()
plt.show()
```
